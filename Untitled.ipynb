```python
import pandas as pd
from zipfile import ZipFile
from io import BytesIO
import matplotlib.pyplot as plt
import os
```

```python
# step 1 : TO extract the dataset

zip_file = 'names.zip'
with ZipFile(zip_file,'r') as zip_ref:
    file_names=zip_ref.namelist()
    print(f"Files in the Zip;{file_names}")
    zip_ref.extractall()
```

```python
# step 2  to load the data set in dataframe 
def load_baby_names(file_name):
    df=pd.read_csv(file_name,names=['Name','Gender','Count'])
    year=int(file_name[3:7])
    df['Year']=year
    return df

all_data = pd.DataFrame()
for file_name in file_names:
    if file_name.startswith('yob')and file_name.endswith('.txt'):
        df=load_baby_names(file_name)
        all_data=all_data._append(df,ignore_index=True)
print(all_data.head())        
```

```python
# step 3 Number of male and female babies borin in particular year
births_by_gender_year = all_data.groupby(['Year', 'Gender']).sum().unstack()
births_by_gender_year['Count'].plot(kind='bar', figsize=(20, 10), width=0.8)
plt.title('Number of Male and Female Babies Born Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Babies')
plt.legend(['Female', 'Male'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a countplot for the 'Gender' column
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Gender', data=all_data)

# Add labels on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()),  # X and Y positions
                ha='center', va='baseline',  # Horizontal and vertical alignment
                fontsize=12, color='black', xytext=(0, 5),  # Text properties
                textcoords='offset points')  # Offset to place the text slightly above the bar

# Adding title and labels
plt.title('Number of Male and Female Babies')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Step 4: Popular baby names 
popular_names = all_data.groupby('Name')['Count'].sum().sort_values(ascending=False)
top_100_names=popular_names.head(100)
top_100_names
```

```python
# Step 5: Visualize the top 10 baby names
top_10_names = top_100_names.head(10)
top_10_names.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Top 10 Popular Baby Names')
plt.xlabel('Names')
plt.ylabel('Number of Babies')
plt.show()
```
